# Convergence testing for o-B14

Energy versus parameters


```python
# Prework

from vmatplot.convergence import *
import matplotlib.pyplot as plt

kpoints_boundary = (40, None)
encut_boundary = (400, None)
```

```python
# 1.1 Energy versus K-points

summarize_parameters("1.1_convergence_o-B14/energy_kpoints_480")
kpoints_450_list = ["ENCUT=450", "1.1_convergence_o-B14/energy_kpoints_450/energy_parameters.dat", kpoints_boundary, "green"]
kpoints_480_list = ["ENCUT=480", "1.1_convergence_o-B14/energy_kpoints_480/energy_parameters.dat", kpoints_boundary, "blue"]
kpoints_list = [kpoints_450_list, kpoints_480_list]

plot_energy_parameters("Energy versus K-points for o-B₁₄", "kpoints", kpoints_list)
plt.savefig("figures/1.1_energy_kpoints_o-B14.pdf")
```

```python
# 1.2 Cohesive energy versus K-points

summarize_cohesive_energy("1.1_convergence_o-B14/energy_kpoints_480", "1.1_convergence_o-B14/cohesive_kpoints_480")
cohesive_kpoints_480 = ["ENCUT=480", "1.1_convergence_o-B14/cohesive_kpoints_480/cohesive_energy.dat", kpoints_boundary, "blue"]
cohesive_kpoints = [cohesive_kpoints_480]

plot_cohesive_energy_parameters("Energy versus energy cutoff for o-B₁₄", "kpoints", cohesive_kpoints)
plt.savefig("figures/1.2_cohesive_energy_kpoints_o-B14.pdf")
```

```python
# 1.3 Energy versus energy cutoff

summarize_parameters("1.1_convergence_o-B14/energy_encut_1260")
summarize_parameters("1.1_convergence_o-B14/energy_encut_8721")
encut_1260_list = ["K-points:10,14,9", "1.1_convergence_o-B14/energy_encut_1260/energy_parameters.dat", encut_boundary, "purple"]
encut_8721_list = ["K-points:19,27,17", "1.1_convergence_o-B14/energy_encut_8721/energy_parameters.dat", encut_boundary, "violet"]
encut_list = [encut_1260_list, encut_8721_list]

plot_energy_parameters("Energy versus energy cutoff for o-B₁₄", "encut", encut_list)
plt.savefig("figures/1.3_energy_encut_o-B14.pdf")
```

```python
# 1.4 Cohesive energy versus energy cutoff

summarize_cohesive_energy("1.1_convergence_o-B14/energy_encut_1260","1.1_convergence_o-B14/cohesive_encut_1260")
cohesive_encut = ["K-points:10,14,9", "1.1_convergence_o-B14/cohesive_encut_1260/cohesive_energy.dat", encut_boundary, "violet"]

# plot_cohesive_energy_parameters("Cohesive energy versus energy cutoff","encut",[cohesive_encut])
plot_cohesive_energy_encut("Cohesive energy versus energy cutoff for o-B₁₄", cohesive_encut)
plt.savefig("figures/1.4_cohesive_energy_encut_o-B14.pdf")
```

```python
# 1.5 Energy versus both k-points and energy cutoff

plot_energy_parameters("Energy versus k-points and energy cutoff for o-B₁₄", ["kpoints", "encut"], kpoints_list, encut_list)
plt.savefig("figures/1.5_energy_kpoints_encut_o-B14.pdf")
```

```python
# 1.6 Cohesive energy versus both k-points and energy cutoff

plot_cohesive_energy_parameters("Cohesive energy versus k-points and energy cutoff for o-B₁₄", ["kpoints", "encut"], cohesive_kpoints, cohesive_encut)
plt.savefig("figures/1.6_cohesive_energy_kpoints_encut_o-B14.pdf")
```
