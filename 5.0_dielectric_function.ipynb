# Dielectric function

```python
# Prework

from vmatplot.dielectric_function import plot_dielectric_function, plot_dielectric_function_spectrum
import matplotlib.pyplot as plt
```

```python
# Dielectric function

dielectric_list = ["", "5.1_dielectric_function/o-B14_n128_k34", "blue"]

plot_dielectric_function("Dielectric function", dielectric_list, [{"xx":"in-plane"},{"zz":"out-of-plane"}], "hor", True, "eV", (1.24, 32))
plt.savefig("figures/5.0_dielectric_o-B14.pdf")
```

```python
# Dielectric function for nbands

dielectric_list = [["(32 nbands)", "5.1_dielectric_function/o-B14_n32_k10", "red",      "solid", 3.2, 1.0],
                   ["(64 nbands)", "5.1_dielectric_function/o-B14_n64_k10", "orange",   "solid", 2.8, 1.0],
                   ["(128 nbands)", "5.1_dielectric_function/o-B14_n128_k10", "green",  "solid", 2.4, 1.0],
                   ["(256 nbands)", "5.1_dielectric_function/o-B14_n256_k10", "blue",   "solid", 2.0, 1.0]]

plot_dielectric_function("Dielectric function versus NBANDS for o-B₁₄", dielectric_list, [{"xx":"in-plane"},{"zz":"out-of-plane"}], "hor", True, "eV", (1.24, 32))
plt.savefig("figures/5.1_dielectric_o-B14.pdf")
```

```python
# Dielectric function for k-points

dielectric_list = [["(10  14  9)", "5.1_dielectric_function/o-B14_n128_k10",  "green"   ],
                   ["(20  28  18)", "5.1_dielectric_function/o-B14_n128_k20", "red"     ],
                   ["(26  37  24)", "5.1_dielectric_function/o-B14_n128_k26", "orange"  ],
                   ["(30  42  28)", "5.1_dielectric_function/o-B14_n128_k30", "yellow"  ],
                   ["(32  45  29)", "5.1_dielectric_function/o-B14_n128_k32", "green"   ],
                   ["(34  48  31)", "5.1_dielectric_function/o-B14_n128_k34", "blue"    ]]

plot_dielectric_function("Dielectric function versus k-points for o-B₁₄", dielectric_list, [{"xx":"in-plane"},{"zz":"out-of-plane"}], "hor", True, "eV", (1.24, 32))
plt.savefig("figures/5.2_dielectric_o-B14.pdf")
```
